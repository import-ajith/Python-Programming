```python
def move(source,target):
    print("Move Disc from {} to {}   ".format(source,target))
```

```python
move("A","B")
```

```python
def hanoi(n,source,helper,target):
    if n==0:
        pass
    else:
        hanoi(n-1,source,helper,target)
        move(source,target)
        hanoi(n-1,helper,source,target)
    
```

```python
hanoi(4,"A","B","C")
```
